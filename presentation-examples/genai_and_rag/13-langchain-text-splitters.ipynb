```python
from langchain_community.document_loaders import WikipediaLoader
```

```python
documents = WikipediaLoader(query="nashville").load()
len(documents)
```

```python
content = documents[0].page_content[900:3000]
print(content)
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0, separator="")
```

```python
text_splitter.create_documents([content])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# RecursiveCharacterTextSplitter attempts to keep larger units of text intact,
# such as paragraphs. If the chunk exceeds the size of the paragraph, then
# it will try to split at the sentence level, and if it still exceeds
# move to the word level.
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
```

```python
text_splitter.create_documents([content])
```

```python
from markitdown import MarkItDown

md = MarkItDown()

content_converted_to_markdown = md.convert(
    "https://en.wikipedia.org/wiki/Nashville,_Tennessee"
).text_content
```

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter

headers_to_split_on = [("#", "Header 1"), ("##", "Header 2"), ("###", "Header 3")]

markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
```

```python
markdown_splitter.split_text(content_converted_to_markdown)
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings

text_splitter = SemanticChunker(OpenAIEmbeddings(model="text-embedding-3-small"))
```

```python
print(text_splitter.create_documents([content]))
```
